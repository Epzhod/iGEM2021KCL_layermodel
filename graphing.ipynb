```python
from calcer import LayerCalc as calc
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# syntax for calling the calculator
calc(10).print_output()
```

```python
x_array = np.arange(1, 3000 + 1, 1)
print(x_array) 
```

```python
y_array = np.array([])
for ele in x_array:
    y_array = np.append(y_array, [calc(ele).protein_grams])
print(y_array)
```

```python
sns.lineplot(x_array, y_array)
```

```python
const_grams = calc(10).protein_grams / 10
const_moles = calc(10).protein_moles / 10
const_units = calc(10).protein_units / 10
const_area  = calc(10).surface_area / 10

print(const_grams, const_moles, const_units, const_area)
```

```python
const_array = np.array([])
for ele in x_array:
    const_array = np.append(const_array, (ele-1)*const_area + calc(ele).ENDS)
print(const_array)

check_array = np.array([])
for ele in x_array:
    check_array = np.append(check_array,  [calc(ele).surface_area])
print(check_array)
```

```python
for ele in (const_array, check_array):
    sns.lineplot(x = x_array, y = ele)
```
